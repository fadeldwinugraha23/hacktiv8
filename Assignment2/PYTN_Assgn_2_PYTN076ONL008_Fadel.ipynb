```python
a_mean_Manhattan.rename(columns={'RESIDENTIAL UNITS':'RESIDENTIAL_UNITS', 'TOTAL UNITS':'TOTAL_UNITS'}, inplace=True)
a_mean_Manhattan.sort_values(by='TOTAL_UNITS', ascending=True, inplace=True)
a_mean_Manhattan
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(410112)
# Real counts data to check with
df = pd.DataFrame({'observed_scores': np.random.randint(1, 100, 40000)})

# Aggregated to value_counts, which is what you have access to
df1 = df.groupby('observed_scores').size().to_frame('counts').reset_index()
```

```python
def weighted_kde(y, weights, bw_method=None):
    sample_range = np.nanmax(y) - np.nanmin(y)
    ind = np.linspace(
        np.nanmin(y) - 0.5 * sample_range,
        np.nanmax(y) + 0.5 * sample_range,
        1000)

    gkde = stats.gaussian_kde(y, bw_method=bw_method, weights=weights)

    y = gkde.evaluate(ind)
    
    return ind, y
```

```python
bw_method=8

fig, ax = plt.subplots()
# Underlying data, with pandas kde
a_mean_Manhattan['RESIDENTIAL_UNITS'].plot.density(ax=ax, bw_method=bw_method, label='RESIDENTIAL UNITS DENSITY', lw=2)

# From aggregated counts data
ind, y = weighted_kde(a_mean_Manhattan['RESIDENTIAL_UNITS'], a_mean_Manhattan['TOTAL_UNITS'], bw_method=bw_method)
ax.plot(ind, y, label='Total Units Manual Weighted KDE', lw=2, linestyle='--')

ax.legend()
plt.show()
```

Persebaran data terbanyak berada di nilai 0 untuk unit residensial.

# Intervals

```python
import pandas as pd
import numpy as np
import math
```

```python
a_ci = df.loc[:, ['BOROUGH', 'TOTAL UNITS']]
a_ci
```

```python
stats = a_ci.groupby(['BOROUGH'])['TOTAL UNITS'].agg(['mean', 'count', 'std'])
print(stats)
print('-'*30)

ci95_hi = []
ci95_lo = []

for i in stats.index:
    m, c, s = stats.loc[i]
    ci95_hi.append(m + 1.96*s/math.sqrt(c))
    ci95_lo.append(m - 1.96*s/math.sqrt(c))

stats['ci95_hi'] = ci95_hi
stats['ci95_lo'] = ci95_lo
pd.DataFrame(stats)
```

Bronx memiliki interval keyakinan 95% terbesar sebesar 4 penjualan unit (high) dan 3 penjualan unit (low)

# Hypotesis Testing

```python
from scipy.stats import ttest_ind
a_HTRU = a_mean_Manhattan['RESIDENTIAL_UNITS']
a_HTTU = a_mean_Manhattan['TOTAL_UNITS']
stat, p = ttest_ind(a_HTRU, a_HTTU)
print('P Value is :',p)
if p > 0.05:
    print('same distribution')
else:
    print('different distributions')
```

Nilai Probabilitas sebesar 0,15 menandakan bahwa hubungan Residential Units dan Total Units mempunyai distribusi yang sama
